```python
import geemap
import ee
from pprint import pprint
```

```python
ee.Authenticate()
ee.Initialize()
```

```python
map = geemap.Map()
```

```python
rgi_hma = ee.FeatureCollection("projects/xuzhihongqu/assets/ncc_glacier/rgi_hma")

test_era5_land_start_date = '1990-01-01'
test_era5_land_end_date = '1990-01-02'

dataset = ee.ImageCollection('ECMWF/ERA5_LAND/DAILY_AGGR').filterDate(test_era5_land_start_date, test_era5_land_end_date).first()

visualization = {
  'bands': ['temperature_2m'],
  'min': 250,
  'max': 320,
  'palette': [
    '000080', '0000d9', '4000ff', '8000ff', '0080ff', '00ffff',
    '00ff80', '80ff00', 'daff00', 'ffff00', 'fff500', 'ffda00',
    'ffb000', 'ffa400', 'ff4f00', 'ff2500', 'ff0a00', 'ff00ff',
  ]
}

map.addLayer(dataset, visualization, 'ERA5-Land Daily Aggregated')
map.addLayer(rgi_hma, {}, 'RGI HMA')
map.addLayer(rgi_hma.bounds(), {}, 'RGI HMA bounds')
```
